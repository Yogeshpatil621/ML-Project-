```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import AgglomerativeClustering



# Separate features and labels
features = dataset.drop('species', axis=1)  # Replace 'species' with your label column name
labels = dataset['species']

# Convert categorical columns to numerical values if needed
features = pd.get_dummies(features)  # One-hot encode categorical features

# Instantiate Agglomerative Clustering model
ac = AgglomerativeClustering(n_clusters=3)  # Adjust the number of clusters as needed

# Fit and predict clusters
dataset['predict'] = ac.fit_predict(features)

# Show results
print(dataset.head(3))
```

```python
sns.pairplot(data=dataset,hue="predict")
plt.show()
```
