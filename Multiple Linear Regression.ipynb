# Multiple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv(r"multiple_linear_regression_dataset.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(),annot=True)
plt.show()
```

```python
p=dataset.iloc[:,:-1]
q=dataset["income"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
p_train, p_test, q_train, q_test = train_test_split(p, q, test_size=0.2, random_state=45)
lr = LinearRegression()
lr.fit(p_train, q_train)
```

```python
lr.score(p_test,q_test)*100
```

```python
lr.coef_
```

```python
lr.predict(p_test)
```

```python
lr.predict([[47,2]])
```
