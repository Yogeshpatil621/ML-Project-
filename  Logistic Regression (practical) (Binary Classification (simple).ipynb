# Logistic Regression (practical) (Binary Classification)(simple)

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.head(3)
```

```python
dataset.drop(columns=["EstimatedSalary"], inplace=True)
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
ln = LogisticRegression()
ln.fit(x_train, y_train)
```

```python
ln.score(x_test,y_test)*100
```

```python
ln.predict([[19]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y= ln.predict(x),data=dataset, color="red")
plt.show()
```

# Logistic Regression (practical) (Multiclass Classification)

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy  as np
```

```python
dataset=pd.read_csv("IRIS .csv")
dataset.head(3)
```

```python
dataset["species"].unique()
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(data=dataset,hue="species")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["species"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# ovr

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class="ovr")
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test) * 100
```

# Auto

```python
from sklearn.linear_model import LogisticRegression
lr2= LogisticRegression()
```

```python
lr2.fit(x_train,y_train)
```

```python
lr2.score(x_test, y_test) * 100
```
