# Naive Bayes (practical)


```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("Naive-Bayes-Classification-Data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="glucose",y="bloodpressure",data=dataset,hue="diabetes")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["diabetes"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sns.kdeplot(data=dataset["glucose"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

# GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(x_train, y_train)
```

```python
gn.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gn)
plt.show()
```

# MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
mn = MultinomialNB()
mn.fit(x_train, y_train)
```

```python
mn.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mn)
plt.show()
```

# BernoulliNB

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
