# Apriori Algorithm (Association Rule) Practical

```python
import pandas as pd 
import numpy as np 
```

```python
dataset=pd.read_csv("groceries - groceries.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
market= []
for i in range (0,dataset.shape[0]):
    cus=[]
    for j in dataset.columns:
        if type(dataset[j][i])==str:
            cus.append(dataset[j][i])
            market.append(cus)
```

```python
l = []
for i in market:
    for j in market:
         l.append(i)  # Correctly using append with parentheses
```

```python
import collections
p = collections.Counter(l)
```

```python
d={"Item Name":p.keys(),"value":()}
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)  
```

```python
tr.transform(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.07,use_colnames=True,max_len=3)
```
