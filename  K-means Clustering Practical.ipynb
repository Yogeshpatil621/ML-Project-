```python
import pandas as pd
import os

# Define the file path
file_path = "IRIS .csv"  # Update this path if needed

# Check if the file exists
if not os.path.isfile(file_path):
    print(f"The file {file_path} does not exist. Please check the file path.")
else:
    # Load the dataset
    dataset = pd.read_csv(file_path)
    print("File loaded successfully.")
    # Proceed with data processing...
```

```python
# Calculate WCSS for different numbers of clusters
wcss = []
for i in range(2, 21):
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(range(2, 21), wcss, marker='o')  # Corrected line
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal Number of Clusters")
plt.xticks(range(2, 21))  # Ensure x-ticks are properly set
plt.show()
```

```python
from sklearn.cluster import KMeans
# Create KMeans instance
kmn = KMeans(n_clusters=3)
# Fit the model and predict clusters
dataset["predict"] = kmn.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset,hue="predict")
plt.show()
```
