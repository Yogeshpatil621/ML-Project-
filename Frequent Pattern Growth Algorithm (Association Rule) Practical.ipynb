# Frequent Pattern Growth Algorithm (Association Rule) Practical

```python
import pandas  as pd 
```

```python
dataset=pd.read_csv("groceries - groceries.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
market= []
for i in range (0,dataset.shape[0]):
    cus=[]
    for j in dataset.columns:
        if type(dataset[j][i])==str:
            cus.append(dataset[j][i])
            market.append(cus)
```

```python
market
```

```python
import time

ll = []
start_time = time.time()
for i in market:
    for j in market:
        l.append(j)
        if len(ll) % 1000 == 0:  # Log progress every 1000 iterations
            print(f"Processed {len(ll)} items in {time.time() - start_time:.2f} seconds")
```
